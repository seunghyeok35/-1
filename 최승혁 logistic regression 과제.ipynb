```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

targetUrl="https://raw.githubusercontent.com/seunghyeok35/-1/main/diabetes.csv"
df2=pd.read_csv(targetUrl)
df2.head()
x= df2.iloc[:, 1: -1].values # 입력값
y= df2.iloc[: , -1].values #클래스
plt.scatter(x[:, 0][y == 0], x[:, 1][y == 0], marker='o',color='red', label='0') # 플랏 뿌려주기
plt.scatter(x[:, 0][y == 1], x[:, 1][y == 1], marker='x',color='blue', label='1')
plt.legend(["class 0", "class 1"], loc=4)
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")

plt.title("1")
plt.show()

```
