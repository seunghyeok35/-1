```python
# 케이글에서 데이터 셋 불러오기
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

targetUrl="https://raw.githubusercontent.com/seunghyeok35/-1/main/auto-mpg.csv"
df2= pd.read_csv(targetUrl)
df2.head()
```

```python
# 무게에 따른 연비 그래프 그려보기
X = df2["weight"]
y = df2["mpg"]
plt.plot(X, y, 'o')
plt.show()
```

```python
# linear regression 코드
line_fitter = LinearRegression()
line_fitter.fit(X.values.reshape(-1,1), y)
```

```python
# 무게가 1700kg (스포티지) 일때의 연비 물어보기
line_fitter.predict([[1700]])
```

```python
# 기울기 물어보기
line_fitter.coef_
```

```python
# x 절편 물어보기
line_fitter.intercept_
```

```python
# linear regression 한 것을 그려보기
plt.plot(X, y, 'o')
plt.plot(X,line_fitter.predict(X.values.reshape(-1,1)))
plt.show()
```
